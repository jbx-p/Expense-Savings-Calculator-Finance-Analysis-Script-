## Expense & Savings Calculator (Finance Analysis, Script)

```python
bills = { #bills is monthly
    "rent": 1200,
    "phone_recharge": 100,
    "internet": 100,
    "electricity": 1500,
    "water": 300,
    "transportation": 300,
    "personal_care": 200,
    "food": 600,
    "entertainment": 400,
    "gym": 250,
    "other": 250,   
}
```

```python
def yearly():
    paycheck = int(input("What is your monthly paycheck amount? ")) #user enters monthly paycheck
    profit = int(input("what is the monthly profit from your side hustles? ")) # user enters monthly profit from side hustles
    
    
    income = paycheck * 12 #this here multiplies the paycheck by 12 for the 12 months in the year
    total_profit = profit * 12
    
    
    gross_income = income + total_profit
    tax = int(input("what is your total tax payed? ")) # tax paid yearly
    
    net_income = gross_income - tax
    print(f"Your yearly net income is: ${net_income}")
    
    total_bills1 = bills
    
    total_bills2 = sum(total_bills1.values())
    cost_of_living = total_bills2 * 12  # multiply by twelve to get yearly total 

    print(f"Your yearly cost of living is: ${cost_of_living}") 
    
    after_col = net_income - cost_of_living
    print(f"Your yearly savings after cost of living is: ${after_col}")
    
    #now that i've determined my cost of living, I'll calculate how much I can save and how much I left to spend for the year.
    # it has been advised to deposit about 20% of one's income to saving. so I'll do 40% of the remaining income after cost of living.
    
    savings = after_col * 0.4
    print(f"Your yearly savings (40% of remaining income) is: ${savings}")
    
    balance = after_col - savings
    
    return print(f"\nMoney left after savings: ${balance}")
yearly()
    
```

```python
# Now that I have an idea of how much i'm spending yearly, let's see how this mothly breakdown can help me manage my expenses 

def monthly():
    monthly_paycheck = int(input("what is your monthly paycheck amount? ")) #user enters monthly paycheck
    monthly_profit = int(input("what is the monthly profit from your side hustles? "))
    
    # user refuses to pay tax once again! we should arrest her! lol
    # tax is 0, so i skip the variable
    
    total_income = monthly_paycheck + monthly_profit
    print(f"Your total monthly income is: ${total_income}")
    
    bills = {
        'rent': int(input("What is your monthly rent amount? ")),
        'phone_recharge': int(input("What is your monthly phone recharge amount? ")),
        'internet': int(input("What is your monthly internet amount? ")),
        'electricity': int(input("What is your monthly electricity amount? ")),
        'water': int(input("What is your monthly water amount? ")),
        'transportation': int(input("What is your monthly transportation amount? ")),
        'personal_care': int(input("What is your monthly personal care amount? ")),
        'food': int(input("What is your monthly food amount? ")),
        'entertainment': int(input("What is your monthly entertainment amount? ")),
        'gym': int(input("What is your monthly gym amount? ")),
        'other': int(input("What is your monthly other expenses amount? ")),
        
    }
    
    bills_list = [(a,b) for a,b in bills.items()] # this converts the dictionary to a list of tuple so user sees the values for every expense.
    print("Here is the breakdown of your monthly bills:  \n", bills_list)
    
    total_bills2 = sum(bills.values())
    
    def bills_ana():
        if total_bills2 > total_income:
            print(f'You are  living above your means!\n Your Monthly Income is ${total_income} and Total expense for the month is ${total_bills2} ')
        else:
            print(f'Total money spent on bils and expenses is: ${total_bills2}')
        return
    bills_ana()
    
    after_col = total_income - total_bills2
    savings = after_col * 0.4
    
    def savings_res():
        if savings > 0:
            print(f'You have saved ${savings} this month!, weldone!')
        else:
            print('You need to cut down on your expenses to save money!')
        return
    savings_res()
    
    #print(f'Total money saved this month is: ${savings}'  )
    
    return
monthly()
```
